### Step0. 환경설정
전처리부터 데이터 예측까지의 과정에서 필요한 라이브러리 환경을 구축합니다.

```python
def install_and_import(package):
    import importlib
    try:
        importlib.import_module(package)
    except ImportError:
        import pip
        pip.main(['install', package])
    finally:
        globals()[package] = importlib.import_module(package)
```
